### Initial and final configurations

```python
filename = "../data/init_conf.dat"
conf_i = pd.read_csv(filename,header=0)
filename = "../data/fin_conf.dat"
conf_f = pd.read_csv(filename,header=0)
```

```python
#plotting

r = 150

plt.figure(figsize=[5,5])
plt.scatter(conf_i['x'].to_list(),conf_i['y'].to_list(),c="black",s=0.5,label="init conf")
plt.scatter(conf_f['x'].to_list(),conf_f['y'].to_list(),c='red'  ,s=0.5,label="fin conf")

circ = plt.Circle((0,0), radius=r, color='r',fill=False,alpha=0.5)
circ2 = plt.Circle((0,0),radius=r*2, color='b',fill=False, alpha=0.5)
circ3 = plt.Circle((0,0),radius=r*3, color='g',fill=False, alpha=0.5)

plt.gca().add_patch(circ)
plt.gca().add_patch(circ2)
plt.gca().add_patch(circ3)
plt.legend()
plt.show() 
```

```python
filename = "../data/atoms_path.dat"
df_atoms_path = pd.read_csv(filename,header=0)
```

```python
x = range(0,len(df_atoms_path['y1'].to_list()))
total = 50
for i_atom in  range(1, total):
    plt.plot(x,df_atoms_path['y'+str(i_atom)].to_list())
plt.plot(x,df_atoms_path['y'+str(total)].to_list(),ls='--',c='black')
plt.show()
```

### Density Profile

```python
filename = "../data/density_profile.dat"
df_density = pd.read_csv(filename,header=0)
```

```python
bins = df_density['rmin'].to_list()
bins.append(df_density.at[df_density.shape[0]-1,"rmax"])
plt.stairs(df_density['dens'].to_list(),edges=bins,fill=True)
# plt.xlim((0,300))
plt.show()
```
